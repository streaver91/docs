```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier
plt.rcParams['figure.figsize'] = (15, 5)
```

```python
data = pd.read_csv('data.csv')
```

```python
data['ln_neg_spdst33'] = np.log(-1.0 * data['spdst33'])
```

```python
has_ln = data.columns.str.contains('ln_')
ln_cols = data.columns[has_ln]
ori_cols = data.columns[~has_ln]
```

```python
data[ln_cols].describe(percentiles=[0.05, 0.1, 0.25, 0.5, 0.75, 0.9, 0.95])
```

```python
data[ori_cols].describe(percentiles=[0.05, 0.1, 0.25, 0.5, 0.75, 0.9, 0.95])
```

```python
bad_cohesive_energy = data['cohesive-energy'] > 6.5
bad_bulk = data['bulk'] > 10
bad_shear = data['shear'] > 17
```

```python
bad_formation_energy = data['formation-energy'] > 10
bad_migration_energy = data['migration-energy'] > 10
bad_relaxation_volume = (data['relaxation-volume'] < 0) | (data['relaxation-volume'] > 10)
```

```python
bad_data = bad_cohesive_energy | bad_bulk | bad_shear | bad_formation_energy | bad_migration_energy | bad_relaxation_volume
```

```python
cleaned_data = data[~bad_data]
```
